```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/twitter_training.csv')  # Adjust the path if needed
df.head()
```

```python
from textblob import TextBlob

# Function to calculate sentiment polarity
def get_sentiment(Positive):
    blob = TextBlob(Positive)
    return blob.sentiment.polarity

# Apply the function to the 'text' column
df['CalculatedSentiment'] = df['Positive'].apply(get_sentiment)

# Classify the sentiment
def classify_sentiment(polarity):
    if polarity > 0:
        return 'positive'
    elif polarity < 0:
        return 'negative'
    else:
        return 'neutral'

df['SentimentLabel'] = df['CalculatedSentiment'].apply(classify_sentiment)

# Show the results
df[['Positive', 'CalculatedSentiment', 'SentimentLabel']].head()
```

```python
from textblob import TextBlob

# Function to calculate sentiment polarity
def get_sentiment(Borderlands):
    blob = TextBlob(Borderlands)
    return blob.sentiment.polarity

# Apply the function to the 'text' column
df['CalculatedSentiment'] = df['Borderlands'].apply(get_sentiment)

# Classify the sentiment
def classify_sentiment(polarity):
    if polarity > 0:
        return 'positive'
    elif polarity < 0:
        return 'negative'
    else:
        return 'neutral'

df['SentimentLabel'] = df['CalculatedSentiment'].apply(classify_sentiment)

# Show the results
df[['Borderlands', 'CalculatedSentiment', 'SentimentLabel']].head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count the occurrences of each sentiment
sentiment_counts = df['SentimentLabel'].value_counts()

# Plot the sentiment distribution
plt.figure(figsize=(8, 5))
sns.barplot(x=sentiment_counts.index, y=sentiment_counts.values, palette='viridis')
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Number of Tweets')
plt.show()
```
