```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data
data = {
    'age': [22, 25, 30, 35, 40, 22, 25, 30, 35, 40, 22, 25, 30, 35, 40],
    'gender': ['Male', 'Female', 'Female', 'Male', 'Female', 'Female', 'Female', 'Male', 'Female', 'Female', 'Male', 'Female', 'Female', 'Male', 'Female']
}

df = pd.DataFrame(data)

# Bar chart for gender distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='gender', data=df)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

# Histogram for age distribution
plt.figure(figsize=(8, 6))
sns.histplot(df['age'], bins=10, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate synthetic data
np.random.seed(0)  # For reproducibility
n = 100  # Number of samples

ages = np.random.randint(18, 70, size=n)
genders = np.random.choice(['Male', 'Female'], size=n)

# Create DataFrame
df = pd.DataFrame({
    'age': ages,
    'gender': genders
})

# Bar chart for gender distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='gender', data=df)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

# Histogram for age distribution
plt.figure(figsize=(8, 6))
sns.histplot(df['age'], bins=10, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate synthetic data
np.random.seed(0)  # For reproducibility
n = 200  # Number of samples

# Create synthetic data
purchase_amounts = np.random.uniform(10, 500, size=n)  # Random purchase amounts between $10 and $500
product_categories = np.random.choice(['Electronics', 'Clothing', 'Home & Garden', 'Beauty', 'Toys'], size=n)
payment_methods = np.random.choice(['Credit Card', 'Debit Card', 'PayPal', 'Gift Card'], size=n)

# Create DataFrame
df = pd.DataFrame({
    'purchase_amount': purchase_amounts,
    'product_category': product_categories,
    'payment_method': payment_methods
})

# Bar chart for product category distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='product_category', data=df)
plt.title('Product Category Distribution')
plt.xlabel('Product Category')
plt.ylabel('Count')
plt.show()

# Bar chart for payment method distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='payment_method', data=df)
plt.title('Payment Method Distribution')
plt.xlabel('Payment Method')
plt.ylabel('Count')
plt.show()

# Histogram for purchase amount distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['purchase_amount'], bins=20, kde=True)
plt.title('Purchase Amount Distribution')
plt.xlabel('Purchase Amount ($)')
plt.ylabel('Frequency')
plt.show()
```
