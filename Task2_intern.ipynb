```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# For displaying plots in the notebook
%matplotlib inline
```

```python
# Set the seed for reproducibility
np.random.seed(42)

# Generate random data
n_samples = 200
data = {
    'Age': np.random.randint(1, 80, n_samples),
    'Fare': np.random.uniform(10, 100, n_samples),
    'Survived': np.random.choice([0, 1], n_samples),
    'Pclass': np.random.choice([1, 2, 3], n_samples),
    'Sex': np.random.choice(['male', 'female'], n_samples),
    'Embarked': np.random.choice(['C', 'Q', 'S'], n_samples)
}

# Create a DataFrame
df = pd.DataFrame(data)

# Display the first few rows of the dataset
df.head()
```

```python
# Check for missing values
print(df.isnull().sum())

# Example: Fill missing values if there were any (this dataset doesn't have missing values)
# df['Age'].fillna(df['Age'].median(), inplace=True)

# Check data types
print(df.dtypes)
```

```python
# Summary statistics
df.describe()
```

```python
# Histograms
df['Age'].plot(kind='hist', bins=20, title='Age Distribution')
plt.xlabel('Age')
plt.show()

df['Fare'].plot(kind='hist', bins=20, title='Fare Distribution')
plt.xlabel('Fare')
plt.show()

# Count plot for categorical data
sns.countplot(x='Pclass', data=df)
plt.title('Passenger Class Distribution')
plt.show()

sns.countplot(x='Sex', data=df)
plt.title('Gender Distribution')
plt.show()

sns.countplot(x='Embarked', data=df)
plt.title('Port of Embarkation Distribution')
plt.show()

```

```python
# Survival rate by Pclass
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival by Passenger Class')
plt.show()

# Box plot for Fare by Pclass
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.title('Fare by Passenger Class')
plt.show()

# Scatter plot for Age and Fare
plt.scatter(df['Age'], df['Fare'])
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter plot of Age vs Fare')
plt.show()
```

```python
# Pivot table for Survival by Pclass and Sex
pivot_table = pd.pivot_table(df, values='Survived', index=['Pclass'], columns=['Sex'], aggfunc=np.mean)
print(pivot_table)

# Heatmap for the pivot table
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu')
plt.title('Survival Rate by Passenger Class and Gender')
plt.show()
```

```python
# Pie chart for Passenger Class (Pclass)
pclass_counts = df['Pclass'].value_counts()
pclass_labels = pclass_counts.index
pclass_sizes = pclass_counts.values

plt.figure(figsize=(8, 8))
plt.pie(pclass_sizes, labels=pclass_labels, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Passenger Class (Pclass)')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Pie chart for Gender (Sex)
sex_counts = df['Sex'].value_counts()
sex_labels = sex_counts.index
sex_sizes = sex_counts.values

plt.figure(figsize=(8, 8))
plt.pie(sex_sizes, labels=sex_labels, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Gender (Sex)')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Pie chart for Port of Embarkation (Embarked)
embarked_counts = df['Embarked'].value_counts()
embarked_labels = embarked_counts.index
embarked_sizes = embarked_counts.values

plt.figure(figsize=(8, 8))
plt.pie(embarked_sizes, labels=embarked_labels, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Port of Embarkation (Embarked)')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
